```python
# QUESTÃO 2 - Soma dos Números

num = int(input("Digite um número: "))
soma = 0
for i in range (1, num+1):
  soma = soma + i
print (soma)
```

```python
# QUESTÃO 3 - Tabuada

num = int(input("Digite um número de 1 a 10: "))
for i in range (1, 11):
  print (num, "x", i, "=", num*i)
```

```python
# QUESTÃO 4 - Fatorial

num = int(input("Digite um número: "))
fatorial = 1

while num < 0:
  num = int(input("Número inválido. Digite um número POSITIVO: "))

for i in range (1, num+1):
  fatorial = fatorial * i
print (fatorial)
```

```python
# QUESTÂO 5 - Caixa Eletrônico

valor = int(input("Digite o valor do saque: "))
notas = [100, 50, 20, 10, 5, 2, 1]
i = 0
while i < len(notas):
    nota = notas[i]
    quantidade = 0
    while valor >= nota:
        valor -= nota
        quantidade += 1
    if quantidade > 0:
        print(f"{quantidade} nota(s) de R${nota}")
    i += 1
```
